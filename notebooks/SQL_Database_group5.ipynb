```python
#Function to clean columns names
def format_columns(df,year):
    df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
    df.insert(0, 'year', year)

    return df
```

```python
df_2018 = format_columns(df_2018, 2018)
df_2019 = format_columns(df_2019, 2019)
df_2020 = format_columns(df_2020, 2020)
```

```python
df_2018 = df_2018.rename(columns={'score': 'ladder_score'})
df_2018.head()
```

```python
df_2019 = df_2019.rename(columns={'score': 'ladder_score'})
df_2019.head()
```

```python
df_2020.head()
```

```python
#sort df_202 by ladder_score to be sure is sorted, restart index and place rank to each country
df_2020 = df_2020.sort_values(by='ladder_score', ascending=False)
df_2020.reset_index(drop=True, inplace=True)
df_2020.insert(1, "rank", range(1, len(df_2020) + 1))
df_2020
```

```python
# columns to drop
#what we do with distopia?
columns_to_keep_2020 = [
    'year',
    'rank',
    'country_name',
    'ladder_score',
    'explained_by:_log_gdp_per_capita',
    'explained_by:_social_support',
    'explained_by:_healthy_life_expectancy',
    'explained_by:_freedom_to_make_life_choices',
    'explained_by:_generosity',
    'explained_by:_perceptions_of_corruption',
    'dystopia_+_residual'
]
df_2020 = df_2020[columns_to_keep_2020]
df_2020.head()
```

```python
results = {
    'year': ['2018', '2019', '2020',],
    'mean': [
        df_2018['ladder_score'].mean(),
        df_2019['ladder_score'].mean(),
        df_2020['ladder_score'].mean()
    ],
    'median': [
        df_2018['ladder_score'].median(),
        df_2019['ladder_score'].median(),
        df_2020['ladder_score'].median()
    ],
    'std': [
        df_2018['ladder_score'].std(),
        df_2019['ladder_score'].std(),
        df_2020['ladder_score'].std()
    ]
}

df_summary = pd.DataFrame(results)
df_summary
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_summary['year'], df_summary['mean'], label='Mean', marker='o')
plt.title('Mean of Score Over Years')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.show()
```
